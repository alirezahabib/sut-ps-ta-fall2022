<div dir = "rtl" style="direction:rtl;line-height:300%;">

پاسخ:<br/>
        <hr/>
        فرض کنید $U$ یک متغیر تصادفی uniform بر بازه ی 0 تا  1 باشد.<br/> اگر $F_1(x)$ تابع توزیع تجمعی $U$ و $F_2(x)$ تابع توزیع تجمعی متغیر نمایی مطلوب ما ($X$) باشد، آنگاه:<br/>
        $F_2(x)=P(X\le x)=1-e^{-x}=u\Rightarrow x=-ln(1-u)\Rightarrow X=-ln(1-U)$
    </font>
</div>

<div dir = "rtl" style="direction:rtl;line-height:300%;">
	<font face="XB Zar" size=5>
                <hr/> 
       50000 نمونه از متغیر تصادفی uniform بین 0 و 1 تولید کرده و تابع بالا را روی آن اعمال کنید. پارامتر توزیع نمایی را دلخواه در نظر بگیرید ولی آن را ذکر کنید. نتیجه را بر روی نمودار نمایش دهید. 
   </font>
</div>
    </font>

```r
#lambda = 1
u <- runif(50000)
e <- sapply(u, function(x) -log(1-x))
print(length(e[e > 1]))
hist(e)
```

10000 نمونه از متغیر تصادفی binomial با استفاده از متغیرهای تصادفی از توزیع نمایی که در مرحله ی قبل به دست آورده اید تولید کرده و بر روی نمودار نمایش دهید.<br/>

```r
# n = 10000, p = 0.1
n = 10000
p = 0.3
d <- sapply(e, function(x) qbinom(pexp(x, rate=1), n, p))
hist(d)
```

```r
f <- 1000
t <- seq(0, 2000, 5)
df <- data.frame(time = t,
                 send = 10 * sin(t/1000000 * 2 * pi * f) + 50,
                 receive1 = (5.4 * (sin(t/1000000 * 2 * pi * f - pi/6) + rnorm(401, 0, 0.06)) + 50),
                 receive2 = (2.1 * (sin(t/1000000 * 2 * pi * f - pi/2) + rnorm(401, 0, 0.06)) + 50),
                 receive3 = (1.3 * (sin(t/1000000 * 2 * pi * f - pi) + rnorm(401, 0, 0.06)) + 50))
write.csv(df,"data.csv")
```

همان‌طور که احتمالا حدس زدید گیرنده‌ها علاوه بر آن که اطلاعات را با مقداری نویز و تضعیف شده دریافت می‌کنند آن را با تاخیر هم می‌بینند.
یعنی نمودار داده‌های گیرنده‌ها در مقایسه با داده‌های فرستنده شیفت خورده است.
به دلیل نویزی که در اطلاعات وجود دارد هیچوقت نقاط با شیفت دادن دقیقا روی هم قرار نمی‌گیرند.
یک روش خوب برای پیدا کردن تاخیر بین این دو موج شیفت دادن یک موج و محاسبه‌ی کواریانس بین این دو است تا جایی که این کواریانس بیشینه شود.
<br>
(اگر نویزی وجود نداشت این مقدار برای داده‌های نرمالایز شده‌ی یک می‌بود چرا که کواریانس یک متغییر نرمالایز شده با خودش یک است )
<br><br>
داده‌های گیرنده‌ی اول را به ازای جابه‌جایی‌های مضرب 5 از صفر تا 1000 میکروثانیه (نصف کل بازه‌ی زمانی) به سمت چپ شیفت دهید. سپس داده‌های 1000 میکروثانیه ابتدایی از فرستنده و گیرنده باقی می‌ماند را جدا کنید و کواریانس این دو را محاسبه کنید.
حال نمودار کواریانس بر حسب شیفت را رسم کنید.
